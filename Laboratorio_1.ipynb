```
#Ejercicio 1: Libras a gramos 
def libras_a_gramos(libras):
	gramos = libras * 453.59237
	return gramos
print(libras_a_gramos(0.), 0.)
print(libras_a_gramos(0.5), 226.796185)
print(libras_a_gramos(1.), 453.59237)
print(libras_a_gramos(1e12), 453592370000000.0)
```

```
#Ejercicio 2: Declinación solar 
import math
def declinacion(dias):
  declinacion = 23.45 * math.sin((2*math.pi*(dias+284))/365)
  return declinacion 
print(declinacion(355), -23.45)
print(declinacion(355), -23.44)
print(declinacion(81), .0)
print(declinacion(1), -23.0117)
print(declinacion(1), -23.0116)
print(declinacion(100), 7.533773566685933)
```

```
#Ejercicio 3: Área del jardín 
from math import pi
def area_jardin(r):
  espacio = math.pi*(r**2) - 2*(r**2)
  return espacio
#print (area_jardin(12))
print(area_jardin(0.), 0.)
print(area_jardin(1.), pi - 2.)
print(area_jardin(2.)/4., pi - 2.)
print(area_jardin(.5)*4., area_jardin(1.))
```

```
#Ejercicio 4: Momento de inercia 
def momento_inercia_tubo(m,a,b):
  inercia = (m*(a**2+b**2))/2
  return inercia 
print(momento_inercia_tubo(0.,2,3), 0.)
print(momento_inercia_tubo(1.,1.,1.5), 1.625)
print(momento_inercia_tubo(2.,1.,1.5), 3.25)
print(momento_inercia_tubo(1.,2.,2.5), 5.125)
```

```
#Ejercicio 5: Cardioide 
import math 
def cardioide(a,theta):
  rho = a*(1+math.cos(theta))
  return rho
print(cardioide(1.,0), 2.0)
print(cardioide(1.,math.pi/3), 1.5)
print(cardioide(1.,2*math.pi/3), 0.5)
print(cardioide(1.,math.pi), 0.)
```

```
#Ejercicio 8: Velocidad de escape
import math 
def velocidad_escape(g,r):
  v = math.sqrt(2*g*r)
  return v 
#print(velocidad_escape(9.8,29000))
print(velocidad_escape(0,1), 0.)
print(velocidad_escape(1,0), 0.)
print(velocidad_escape(1,2), 2.)
print(velocidad_escape(2,4), 4.)
```

```
#Ejercicio 9: Velocidad orbital (órbita elíptica)
import math 
G = 6.67*(10**11)
def velocidad_orbital(M,r,a):
  v = math.sqrt(2*G*M*((1/r)-(1/(2*a))))
  return v 
print(velocidad_orbital(.5/G,.5,.5), 1.)
print(velocidad_orbital(.5/G,1,1.5), math.sqrt(2./3.))
print(velocidad_orbital(1./G,1,1), 1.)
print(velocidad_orbital(.5/G,.5,1.5), math.sqrt(5./3.))
```

```
#Ejercicio 10: Catenaria 
import math
def catenaria(x,a):
  y = a*math.cosh(x/a)
  return y
print(catenaria(1.5,1.), 2.352409615243247)
print(catenaria(0.,1.), 1.)
print(catenaria(0.,2.), 2.)
print(catenaria(-1.,1.), 1.5430806348152437)
```
